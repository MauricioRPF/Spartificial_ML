```python
# Use Pandas to read the csv dataaaaaaaaaaaa
import pandas as pd
```

```python
# Storing the CSV data in pandas DataFrame
path_to_csv = 'https://raw.githubusercontent.com/MauricioRPF/Spartificial_ML/main/W2S3_SDSS%20Tabular%20Data.csv' # I had stored my csv data within MyDrive > Program Datasets > SDSS Tabular Data.csv
df = pd.read_csv(path_to_csv)
df
```

```python
# @title class vs g

from matplotlib import pyplot as plt
import seaborn as sns
figsize = (12, 1.2 * len(df['class'].unique()))
plt.figure(figsize=figsize)
classVSg = sns.violinplot(df, x='g', y='class', inner='box', palette='Dark2')
sns.despine(top=True, right=True, bottom=True, left=True)
```

```python
# @title objid vs dec

from matplotlib import pyplot as plt
import seaborn as sns
def _plot_series(series, series_name, series_index=0):
  palette = list(sns.palettes.mpl_palette('Dark2'))
  xs = series['objid']
  ys = series['dec']

  plt.plot(xs, ys, label=series_name, color=palette[series_index % len(palette)])

fig, ax = plt.subplots(figsize=(10, 5.2), layout='constrained')
df_sorted = df.sort_values('objid', ascending=True)
for i, (series_name, series) in enumerate(df_sorted.groupby('class')):
  _plot_series(series, series_name, i)
  fig.legend(title='class', bbox_to_anchor=(1, 1), loc='upper left')
sns.despine(fig=fig, ax=ax)
plt.xlabel('objid')
_ = plt.ylabel('dec')
```

```python
# Check the count of galaxy, star and qasar in our data

# Adjusting the Figure Size
plt.figure(figsize = (10, 10))

''' We can plot directly using pandas '''
# value_counts() gives counts of unique values in the column it is used on
colors = ['blue', 'pink', 'purple']
ax = df['class'].value_counts().plot(kind = 'bar', color = colors) # kind can be (line, hist, bar, box, pie, etc...)

# Using ax (axis) to display the bar_label using containers[0] as thats where the location is of the counts
ax.bar_label(ax.containers[0], color='black', weight='bold')

# Customizing the Plot a bit more
plt.xticks(weight='bold', rotation=0)
plt.title('Visualizing the distribution of Classes in the Data', color='r', weight='bold', fontsize=12)

# Save the Figure and Display the output
plt.savefig('ClassDistribution.png') # You can download the visualization from files section (left panel)
plt.show()
```

### **Scikit-Image**

```python
# Get the imread from skimage's io
from skimage.io import imread
```

```python
# Get the image path from session storage of colab
img_path = '/content/Moon-Pleiades-AirplaneLR.jpg'
```

```python
# Read the image path and convert it to array of pixels
img_array = imread(img_path)
```

```python
# What is the shape of this image?
print(img_array.shape)
```

**Understand the shape of this image_array**:-

* Axis 0:- Rows
* Axis 1:- Cols
* Axis 3:- Channels

* If channels = 3, then it represents RGB (Red, Green, and Blue Channels)

* If channel = 1 (or only 2D array) then it represents a gray scale output.


```python
# Check the image array
print(img_array)
```

```python
# What are minimum and maximum values in this array?
print(f'Min:- {img_array.min()}\nMax:- {img_array.max()}')
```

```python
# This is uint-8 datatype - crosscheck it
img_array.dtype
```

```python
# Visualize the image
plt.imshow(img_array, cmap = 'viridis')
plt.show()
```

```python
# Grayscale output
plt.imshow(img_array[:, :, 2], cmap='gray') # 0 - red, 1 - green, 2 - blue
plt.show()
```

```python
# Slice the moon and the pleiades constellation
constellation = img_array[60:120, 50:140, :]
moon = img_array[185:233, 214:264, :]

# Tuple to store the data for plotting
plot_data = (
    (constellation, moon),
    ('Pleiades', 'Moon'),
    ('blue', 'gray')
)


# Using for loop to loop through the tuple to create subplot
for index, (data, title, color) in enumerate(zip(*(plot_data))):
  plt.subplot(1, 2, index+1)
  plt.imshow(data)
  plt.title(title, color = color)
  plt.axis('off')


# Adjust the layout and display the output
plt.tight_layout()
plt.show()
```
